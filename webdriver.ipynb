```python
#determines the current page in quartzy
def current_page(driver):
    for page in driver.find_element_by_class_name("pagination").find_elements_by_class_name("page-item"):
        if "active" in page.get_attribute("class"):
            return(page.find_element_by_class_name("page-link").get_attribute("href").split('?page')[-1])

def next_page(driver):
    page_items = driver.find_element_by_class_name("pagination").find_elements_by_class_name("page-item")
    driver.execute_script("arguments[0].click();", page_items[-1].find_element_by_class_name("page-link"))
    return(current_page(driver))

def last_page(driver):
    page= driver.find_element_by_class_name("pagination").find_elements_by_class_name("page-item")[-2]
    return(page.find_element_by_class_name("page-link").get_attribute("href").split('?page')[-1])

def go_to_first_page(driver):
    ## addo functionality for doing nothing if already on first page
    page_items= driver.find_element_by_class_name("pagination").find_elements_by_class_name("page-item")
    driver.execute_script("arguments[0].click();", page_items[1].find_element_by_class_name("page-link"))
    
def go_to_last_page(driver):
    ## add functionality for doing nothing if already on the last page
    page_items= driver.find_element_by_class_name("pagination").find_elements_by_class_name("page-item")
    driver.execute_script("arguments[0].click();", page_items[-2].find_element_by_class_name("page-link"))
```

```python
counter = 1
while(counter != last_page(driver)):
    plasmids = driver.find_elements_by_class_name("item-row")
    for plasmid in plasmids:
        try:
            attachment = plasmid.find_element_by_class_name("item-row-container").find_element_by_class_name("attachment")
            attachment.find_element_by_tag_name("i")
            file_link = plasmid.find_element_by_class_name("item-row-container").find_element_by_class_name("item-link")
            file_link.click()
            file = driver.find_element_by_class_name("name")
            url = file.get_property("href")
            r = requests.get(url, allow_redirects=True)
            open("./quartzy_download/"+file.text,"wb").write(r.content)
            print(url)
            print("attachment")
            time.sleep(3)
        except:
            print("No attachment")
            time.sleep(2)
            next
    next_page(driver)
    time.sleep(2)
    counter=current_page(driver)
```
